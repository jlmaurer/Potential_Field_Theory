## Visualize the results

```python

# ── 1. Build grid ─────────────────────────────────────────────────────────────
r     = np.linspace(1.0, 2.0, 10)
theta = np.linspace(0.05, np.pi - 0.05, 90)   # colatitude, avoid poles
phi   = np.linspace(0, 2 * np.pi, 180, endpoint=False)

R, THETA, PHI = np.meshgrid(r, theta, phi, indexing='ij')

# ── 2. Smooth scalar field: superposition of low-degree harmonics ─────────────
# This mimics the kind of large-scale structure seen in geophysical fields.
f = (  2.0 * np.cos(THETA)
     + np.sin(THETA)**2 * np.cos(2 * PHI)
     - 0.5 * np.sin(2 * THETA) * np.sin(PHI)
     + 0.3 * np.cos(3 * THETA) )

# ── 3. Gradient then divergence (= Laplacian) ─────────────────────────────────
Fr, Ftheta, Fphi = gradient_spherical(f, r, theta, phi)
laplacian = divergence_spherical(Fr, Ftheta, Fphi, r, theta, phi)
```

```python
# ── 4a. 2-D lat/lon maps on the middle shell ──────────────────────────────────
ir = len(r) // 2                          # index of middle shell
lat = 90 - np.degrees(theta)             # convert colatitude → latitude
lon = np.degrees(phi)

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

for ax, data, title in zip(
        axes,
        [f[ir], laplacian[ir] if laplacian is not None else np.full_like(f[ir], np.nan)],
        [r'Scalar field $f$', r'$\nabla^2 f = \nabla \cdot \nabla f$']):
    im = ax.pcolormesh(lon, lat, data, cmap='RdBu_r', shading='auto')
    plt.colorbar(im, ax=ax, shrink=0.85)
    ax.set_xlabel('Longitude (°)')
    ax.set_ylabel('Latitude (°)')
    ax.set_title(f'{title}  (r = {r[ir]:.2f})')

plt.tight_layout()
plt.show()
```

```python
# ── 4b. 3-D sphere with field mapped as colour ────────────────────────────────
from mpl_toolkits.mplot3d import Axes3D   # noqa: F401
from matplotlib.colors import Normalize
from matplotlib import cm

THETA2, PHI2 = np.meshgrid(theta, phi, indexing='ij')
Xs = np.sin(THETA2) * np.cos(PHI2)
Ys = np.sin(THETA2) * np.sin(PHI2)
Zs = np.cos(THETA2)

fig2, axes2 = plt.subplots(1, 2, figsize=(13, 5),
                            subplot_kw={'projection': '3d'})

datasets = [
    (f[ir],       r'$f$',              'RdBu_r'),
    (laplacian[ir] if laplacian is not None else np.full_like(f[ir], np.nan),
     r'$\nabla^2 f$', 'PuOr_r'),
]

for ax, (data, title, cmap_name) in zip(axes2, datasets):
    norm  = Normalize(vmin=np.nanmin(data), vmax=np.nanmax(data))
    fcolors = cm.get_cmap(cmap_name)(norm(data))
    ax.plot_surface(Xs, Ys, Zs, facecolors=fcolors,
                    rstride=1, cstride=1, linewidth=0, antialiased=False)
    ax.set_title(title, pad=8)
    ax.set_box_aspect([1, 1, 1])
    ax.axis('off')

    # add a colourbar using a dummy ScalarMappable
    sm = cm.ScalarMappable(cmap=cmap_name, norm=norm)
    sm.set_array([])
    fig2.colorbar(sm, ax=ax, shrink=0.55, pad=0.02)

plt.suptitle(f'Fields on spherical shell  r = {r[ir]:.2f}', y=1.01)
plt.tight_layout()
plt.show()
```
